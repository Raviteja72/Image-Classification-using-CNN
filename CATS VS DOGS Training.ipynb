```python
DIRECTORY = r'/Users/apple/Downloads/dogscats/dogscats'
CATEGORY = ['cats','dogs']
```

```python
IMG_SIZE = 100

data = []


for category in CATEGORY:
    folder = os.path.join(DIRECTORY, category)
    label = CATEGORY.index(category)
    for img in os.listdir(folder):
        img_path = os.path.join(folder, img)
        img_arr = cv2.imread(img_path)
        img_arr = cv2.resize(cv2.imread(img_arr,cv2.IMREAD_GRAYSCALE), (IMG_SIZE, IMG_SIZE))
        data.append([img_arr, label])
        
```

```python
PATH = r'/Users/apple/Desktop/test/'
CATEGORIES = ['cats','dogs']
```

```python
test_data = []

for cate in CATEGORIES:
    folder1 = os.path.join(PATH, cate)
    label1 = CATEGORIES.index(cate)
    for img1 in os.listdir(folder1):
        img1_path = os.path.join(folder1, img1)
        img1_arr = cv2.resize(cv2.imread(cv2.imread(img1_path),cv2.IMREAD_GRAYSCALE, (IMG_SIZE, IMG_SIZE)))
        test_data.append([img1_arr,label1 ])
    random.shuffle(test_data)
    
```

```python
X_dl = []
y_dl = []
for feat,labe in test_data:
    X_dl.append(feat)
    y_dl.append(labe)
```
